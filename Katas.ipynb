```python
def frecuencia_letras(texto):
    frecuencias = {}

    for letra in texto:
        if letra != " ":  # ignorar espacios
            letra = letra.lower()  # opcional: contar mayúsculas y minúsculas como la misma letra
            
            if letra in frecuencias:
                frecuencias[letra] += 1
            else:
                frecuencias[letra] = 1

    return frecuencias
```

```python
texto = "Soy Alma"
resultado = frecuencia_letras(texto)
print(resultado)
```

```python
def evaluar_notas(lista, nota_aprobado=5):
    media = sum(lista) / len(lista)
    
    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"
        
    return media, estado
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
class Arbol:
    
    def __init__(self):
        # 1. Tronco inicial = 1 y lista de ramas vacía
        self.tronco = 1
        self.ramas = []
    
    
    # 2. Crecer el tronco en 1 unidad
    def crecer_tronco(self):
        self.tronco += 1
    
    
    # 3. Añadir nueva rama de longitud 1
    def nueva_rama(self):
        self.ramas.append(1)
    
    
    # 4. Crecer todas las ramas en 1 unidad
    def crecer_ramas(self):
        self.ramas = [rama + 1 for rama in self.ramas]
    
    
    # 5. Eliminar una rama por posición
    def quitar_rama(self, posicion):
        if 0 <= posicion < len(self.ramas):
            self.ramas.pop(posicion)
        else:
            print("Posición no válida")
    
    
    # 6. Mostrar información del árbol
    def info_arbol(self):
        return {
            "Longitud del tronco": self.tronco,
            "Número de ramas": len(self.ramas),
            "Longitudes de las ramas": self.ramas
        }
```

```python
class UsuarioBanco:
    
    # 1. Inicializar usuario
    def __init__(self, nombre, saldo, cuenta_corriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente
    
    
    # 2. Retirar dinero
    def retirar_dinero(self, cantidad):
        if not self.cuenta_corriente:
            raise Exception(f"{self.nombre} no tiene cuenta corriente.")
        
        if cantidad > self.saldo:
            raise Exception(f"{self.nombre} no tiene saldo suficiente.")
        
        self.saldo -= cantidad
        print(f"{self.nombre} ha retirado {cantidad}. Saldo actual: {self.saldo}")
    
    
    # 3. Transferir dinero desde otro usuario al actual
    def transferir_dinero(self, otro_usuario, cantidad):
        if not otro_usuario.cuenta_corriente or not self.cuenta_corriente:
            raise Exception("Uno de los usuarios no tiene cuenta corriente.")
        
        if cantidad > otro_usuario.saldo:
            raise Exception(f"{otro_usuario.nombre} no tiene saldo suficiente para transferir.")
        
        otro_usuario.saldo -= cantidad
        self.saldo += cantidad
        
        print(f"{otro_usuario.nombre} ha transferido {cantidad} a {self.nombre}")
    
    
    # 4. Agregar dinero al saldo
    def agregar_dinero(self, cantidad):
        self.saldo += cantidad
        print(f"{self.nombre} ha agregado {cantidad}. Saldo actual: {self.saldo}")
```

```python
# 1. Crear usuarios
alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 50, True)

# 2. Agregar 20 al saldo de Bob
bob.agregar_dinero(20)

# 3. Transferir 80 desde Bob a Alicia
alicia.transferir_dinero(bob, 80)

# 4. Alicia retira 50
alicia.retirar_dinero(50)
```

```python
def contar_palabras(texto):
    palabras = texto.lower().split()
    resultado = {}

    for palabra in palabras:
        if palabra in resultado:
            resultado[palabra] += 1
        else:
            resultado[palabra] = 1

    return resultado
```

```python
def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    return texto.replace(palabra_original, palabra_nueva)
```

```python
def eliminar_palabra(texto, palabra):
    palabras = texto.split()
    palabras = [p for p in palabras if p != palabra]
    return " ".join(palabras)
```

```python
def procesar_texto(texto, opcion, *args):

    if opcion == "contar":
        return contar_palabras(texto)

    elif opcion == "reemplazar":
        return reemplazar_palabras(texto, args[0], args[1])

    elif opcion == "eliminar":
        return eliminar_palabra(texto, args[0])

    else:
        return "Opción no válida"
```

```python
texto = "python es fácil y python es potente"

# Contar palabras
print(procesar_texto(texto, "contar"))

# Reemplazar palabra
print(procesar_texto(texto, "reemplazar", "python", "Java"))

# Eliminar palabra
print(procesar_texto(texto, "eliminar", "es"))
```

```python
import math

def calcular_area(figura, datos):
    
    if figura == "rectangulo":
        base, altura = datos
        return base * altura
    
    elif figura == "circulo":
        radio = datos[0]
        return math.pi * radio ** 2
    
    elif figura == "triangulo":
        base, altura = datos
        return (base * altura) / 2
    
    else:
        return "Figura no válida"
```

```python
print(calcular_area("rectangulo", (2, 7)))  
print(calcular_area("circulo", (5,)))        
print(calcular_area("triangulo", (8, 4)))     
```
